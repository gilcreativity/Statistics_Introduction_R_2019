# 단순선형회귀모형(simple linear regression model)

## 모델

* 모델 피팅
   * 추정(estimation)
   * 검증(evaluation)
   
* 단순(simple)은 설명변수(회귀변수; regressonr variable, x)가 한개인 경우

## 모형모수에 대한 가정과 해석(Assumptions and Interpretation of model paramters)
* $E(y_i) = \beta_0 + \beta_i x_i $
* 회귀모형의 절편(intercept): $\beta_0$ 
   * 해석(interpretation)의 여지가 제한된 회귀항(regression term)
   * 제한되 자료 x에 의해 계산되기 때문에 x의 범위에 매우 의존적
* 기울기(slope):  $\beta_1$ :  $x_1의 단위증가당 E(y)의 변화$

## 모델의 기본 가정(underlying assumptions)
* 정규성 이론 가정(Normal Theory Assumption): 모형의 오차가 가우스 분포(Gaussian distribution)을 따름

* 중심화모형(centered model): $y_i = \beta_0^{*} + \beta_1(x-\bar x) + \varepsilon_i, (i = 1, 2, \cdots, n)$

# 모델의 추정 방법

## 1. 최소제곱법(least Squares method)

#### 가정
* $x_{i}$는 비임의(nonrandom)의 오차가 없는 관측치(observation)
* $\varepsilon_{i}$는 평균(mean)이 0, 분산(variance)이 $\sigma^2$이 일정한 확률변수(random variable)
   * 동분산 가정(homogenous variance assumption)
   * 모집단 평균을 나타내는 기대연산자(expectation operator)를 사용하면 평균 0, 분산일정
* $\varepsilon_{i}$은 관측치(observation)간에 서로 상관되어(correlated)있지 않음 가정

#### 단순선형회귀식
* $E(y_i) = \beta_0 + \beta_i x_i $
* $E(y|x) = \beta_0 + \beta_i x_i $ y의 조건부평균(mean y conditional on a specific value of x)


### 잔차제곱합(residual sum of squares)
* $\sum_{i=1}^N (y_i - \hat{y_i})^2$을 최소로 하는 $\beta_i$의 추정량(estimator)
* 잔차제곱합으로 추정된 $\beta_i$의 특성
   * 적합오차(error in fit)인 잔차가 작아지면 적합된 회귀선(fitted regression line)에 가까워짐

## 2. 최대우도법(Maximum Likelihood Estimation)

* 오차항의 확률변수에 대한 분포를 가정할 경우 좋은 점근적 특성(outstanding asymtotic property)을 가지는 최대우도법 활용 가능

#### 가정
* 오차항이 정규분포 따른다고 가정

# 다중선형회귀모형(The Multiple Linear Regression Model)

## 모형
* 하나 이상의 설명변수(회귀변수: regressor variable)을 활용하여 모형 구축(model building)하는 경우
* 종속변수(y)가 설명변수와 거의 선형적 관계가 있는 경우

## 모형의 가정
* 오차의 평균이 0
* 분산이 일정
* 관측값(observation)마다 무상관(uncorrelated)
* 설명변수(X) 는 비임의(nonrandom)이며 무시할 수 있을 정도로 오차가 작은(negligible error)을 관측값으로 가정

## 모형의 추정법으로 최소제곱 과정(least squares procedure) 등을 적용


### 변수의 변환(transformation)

* 정규분포화
* 로그변환(log transformation)
* 역수변환(reciprocal transformation)
* 지수변환(exponential transformation)

# 로지스틱 회귀

### 로짓모형 사용하는 이유
* 변수가 명목 또는 서옇형 척도일 때
    * 종속변수가 이산형(discrete variable)인 경우 

### 로짓 모형 개요

* 로지스틱 분포(logistic distribution)

* 로지스틱 함수 $y = {e^x \over 1+ e^x}$

* 오즈(odds): 승산비(${P\over {1-P}}$)

* 로짓(logit): 오즈(odds)의 로그값

```r
x <- seq(-5, 5, length = 200)
y = 1/sqrt(2*pi)*exp(-x^2 /2)
plot(x, y, type = "l", lwd = 2, col = "blue")
```

```r
x <- seq(-5, 5, length = 200)
y <- dnorm(x, mean = 0, sd =1)
plot(x, y, type = "l", lwd = 2, col = "red")
```

```r
x <- seq(-4, 4, length = 200)
y <- dnorm(x, mean = 0, sd =1)
plot(x, y, type = "l", lwd = 2, col = "red")

x <- seq(-1.64, 1.64, length = 200)
y <- dnorm(x, mean = 0, sd =1)
polygon(c(-1.64, x, 1.64), c(0, y, 0), col = "blue")
```

* linearizable regresson model: 포괄적으로 봤을 때 선형 회귀 모델
  * 회귀계수를 기준으로 판단
  * 설명변수(x)와 종속변수(y)를 적절히 변환할 경우 모두 선형 회귀식으로 표현 가능 : log, e 활용
  * 파라미터를 추정할 때도 일반적인 선형 회귀 모델과 동일한 기법사용
  * 일반화 선형 모델(Generalized linear model)의 특수한 케이스
  * logit transformation 활용
  
$logit(E[Y_i | x_{1, i},\cdots ,x_{m, i}]) = ln  {{p_i}\over{1- p_i}} = \beta \mathbf X_i $

$p_i = logit^{-1}(\beta \mathbf X_i ) = {1 \over {1+ e^{-\beta \mathbf X_i}}}$

```r
model_logit <- glm(health_ins ~ logwage + education + age + jobclass, 
    family = binomial("logit"), data = Wage, maxit = 10000)

summary(model_logit)
```

```r
# log-likelihood 활용 신뢰구간
confint(model_logit)
```

```r
# standard errors 활용한 신뢰구간
confint.default(model_logit)
```

```r
set.seed(13)

new_ind <- sample(1:nrow(Wage), 10)
new_ind
new_data <- Wage[new_ind, c("logwage", 'education', 'age', 'jobclass', 'health_ins')]
new_data
```

```r
predict(model_logit, newdata = new_data)
```

```r
predict_prob <- plogis(predict(model_logit, newdata = new_data))
predict_prob
```

```r
threshold <- 0.19
hat <- ifelse(predict_prob > threshold, "2. No", "1. Yes")
```

```r
res <- data.frame(new_data$health_ins, hat)
res
```

```r
table(res)
```

```r
CM <- confusionMatrix(new_data$health_ins, as.factor(hat))
CM
```

```r
CM$byClass["Balanced Accuracy"]
```

```r
test_threshold <- seq(0.15, 0.25, 0.01)
test_threshold
```

```r
for (i in 1:length(test_threshold)){
    
    threshold <- test_threshold[i]
    hat <- ifelse(predict_prob > threshold, "2. No", "1. Yes")
    CM <- confusionMatrix(new_data$health_ins, as.factor(hat))
    print(paste0(test_threshold[i], " 정확도: "))
    print(CM$byClass["Balanced Accuracy"])
}
```

```r
test_threshold <- seq(0.15, 0.3, 0.01)

acc <- c()

for (i in 1:length(test_threshold)){
    
    threshold <- test_threshold[i]
    hat <- ifelse(predict_prob > threshold, "2. No", "1. Yes")
    CM <- confusionMatrix(new_data$health_ins, as.factor(hat))
    acc[i] <- CM$byClass["Balanced Accuracy"]
}
```

```r
plot(test_threshold, acc, type = "l")
```
