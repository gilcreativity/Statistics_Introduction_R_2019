## 정규분포(Normal distribution)
* 데이터가 충분히 많다면 대분 정규분포를 사용한다고 가정(중심극한정리)
  * 표준정규분포: 평균0, 표준편차:1
  * 정규분포: 평균 0+a, 표준편차: b

### 데이터 정규성 검정
* Shapiro-Wilks test : 표본수가 2000 미만인 경우의 데이터
* Kolmogorove-Smirnov test: 표본수가 2000 초과인 데이터에서 적합
* QQ plot(Graphic test): 시각적 분석


* 귀무가설(H0): 정규분포를 따름
* 대립가설(H1): 정규분포를 따르지 않음
* P-value < 0.05이하인 경우 95%신뢰수준에서 데이터셋은 정규분포를 따르지 않음

```r
shapiro.test(rnorm(1000))
```

```r
shapiro.test(rnorm(100, mean = 5, sd = 3))
shapiro.test(runif(100, min = 2, max = 4))
```

```r
hist(rnorm(100))
hist(runif(100))
```

## 독립성 검정(Test of Independence)

```r
library(dplyr)

df_sub <- df %>% select("Survived", "Pclass") %>% 
group_by(Survived) %>% count(Pclass)

df_sub
```

```r
library(reshape)

df_cast <- cast(df_sub, Pclass ~ Survived)
df_cast
```

```r
rownames(df_cast) <- c("1st", "2nd", "3rd")
colnames(df_cast) <- c("del", 's', "not_s")
df2 <- select(df_cast, "s", "not_s")
df2
```

```r
chisq.test(df2)

# H0(귀무가설): 차이가 없다
# H1(대립가설): 차이가 있다
# P value값이 0.05보다 작으므로 비율차이가 있다
```
