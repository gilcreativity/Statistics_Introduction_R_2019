# Probaility
> * 불확실성(uncertainty)을 가진 사건에 대해 자룔르 근거로 가장 객관적으로 합리적으로 확률(probabilty)을 제시하는게 통계학의 역할: 객관적 자료 --> 확률

> * 확률을 정의하기 위해 필요한 정의
> * 실험(experiment): 실행하기 전에는 그 결과를 알 수 없음
> * 표본공간(sample space): 실험을 하였을 때 나타날 수 있는 모든 결과를 모은 집합(S)
> * 단순(단위) 사건(simple event, elementary event): 표본공간의 원소 하나 하나
> * 사건(event): 표본공간의 부분집합

> * 확률(Probability): 표본공간에서 정의된 함수이며, 0과 1상의 값을 가짐
> * 확률의 기본적 성질 : 
>> 1) 임의의 사건 A에 대해서 $0<=P(A)<=1$
>> 2) 표본 공간에 대한 확률은 언제나 $P(S)=1$
> * 확률의 법칙
>> * (1) 집합의 종류
>>> * 합집합(union)
>>> * 교집합(intersection)
>>> * 여집합(complement)
>>> * 배반사건(disjoint event)
>> * (2) 확률의 법칙
>>> * 여사건의 법칙(law of complement): $ P(\bar{A})=1-P(A)$
>>> * 합의 법칙(law of addition): 본페르니 부등식$P(A\cup B)=P(A)+P(B)-P(A\cap B)$ 

# 조건부 확률과 독립(conditional probability and independence)
> * $P(B)>0$인 경우, 사건 B가 주어졌을 때 사건 A의 조건부 확률
> <center> $P(A|B)=\frac{P(A\cup B)}{P(B)}$ </center>
> * 참고) $P(A \cup B) = P(A|B)P(B)$
> * 참고) $P(A \cup B) = P(B|A)P(A)$
> * 두사건 A와 B가 $ P(A|B)=P(A)$를 만족하면 두사건 A와 B는 서로 독립(independent)이라고 함
> * 두 사건 A와 B는 서로 독립 = $P(A\cup B)=P(A)P(B)$


# 임의표본(random sample)
> * n개의 표본들이 서로 독립(independent)이고 같은 분포(identically distributed)를 가지면 임의표본(random sample)이라고 함
> * 임의 표본은 흔히 iid(independent and identically distributed)라고도 함

# 확률분포(Probability distribution)
> * 확률변수(random variable): 표본공간에서 정의된 함수
>> 1. 이산확률변수(discrete random variable) : 확률변수 X가 취할수 있는 값이 유한개(finite) 또는 자연수와 일대일 대응이 되는 무한 개 일 경우(countable)
>>> * 확률변수 X가 취할수 있는 값을 x1,.., xn 라고 하면 $f(x_i)= P(X=x_i), i = 1,...,k$ 를 확률분포(probability distribution)라 함
>>> * 확률분포는 두가지 성질 만족
>>> <center>$f(x_i)>0, i = 1,...,k$</center>
>>> <center>$\sum_{i_1}^{k}{f({x_i})=1}$</center>
>> 2. 연속확률변수(continuous random variable)
>>> * 확률밀도함수(pdf: probability density function)
>>> <center>$f(x)>=0$</center>
>>> <center>$\int_{-\infty}^{\infty}f(x)\,dx=1$</center>
> * 확률밀도함수의 형태 
>> * 균일 모양(unifrom)
>> * 종 모양(bell-shaped)
>> * 왼편으로 긴꼬리 모양(skewed to the left)
>> * 오른편으로 긴 꼬리 모양 (skewed to the right)

# 기대값(expectation, expected value)
> $E(X) = \begin{cases} \sum_{i=1}^{k}{x_i}{f(x_i)} & \mbox{: }\mbox{이산형 변수인 경우} \\ 
\int xf(x)\,dx & \mbox{: }\mbox{연속형 변수인 경우}\end{cases}$
> * X의 기대값 = $\mu$ : 모평균(population mean)
> * 기대값의 선형성(linearity property of expectation): 상수 a와 b에 대해서서 $E(a + bX) = a+bE(x)$

> * 일반적으로 확률변수 X의 k차 적률(k-th moment)
> $E(X^k) = \begin{cases} \sum_{i=1}^{k}{x_i}^k{f(x_i)} & \mbox{: }\mbox{이산형 변수인 경우} \\ 
\int {x^k}f(x)\,dx & \mbox{: }\mbox{연속형 변수인 경우}\end{cases}$

> * 확률변수 X의 k차 중심적률(k-th central moment)
> $E[(X-\mu)^k] = \begin{cases} \sum_{i=1}^{k}({x_i}-\mu)^k{f(x_i)} & \mbox{: }\mbox{이산형 변수인 경우} \\ 
\int {(x-\mu)^k}f(x)\,dx & \mbox{: }\mbox{연속형 변수인 경우}\end{cases}$

> * 2차중심적률 $E[(X-\mu)^2]\equiv \sigma^2$: 모분산(population variance)
> * 기대값의 선형성을 이용:
>>  $\begin{matrix}\sigma^2 &=& E[(X-\mu)^2] \\ &=& E[(X^2 - 2\mu X+\mu^2)]\\&=& E(X^2)-2\mu E(X) + \mu^2  \\ &=& E(X^2)-\mu^2\end{matrix}$
> * $모분산 = 2차적률-1차적률^2$ 으로 해석 가능

> * $\mu$와 $\sigma^2$ (: unknown constants)은 모집단의 평균과 분산을 나타내며 이는 모수(parameter)로서 알구 없는 값
> * $\mu$와 $\sigma^2$을 알 수 없기 때문에 표본을 추출하여 표본평균과 표본분산을 계산

|      |           모집단          |      |     표본                                   |
|----|----------------------|----|-----------------------------------------|
| 모평균 | $\mu=E(X)$             |표본평균|$\bar{X}=\frac{1}{n}\sum{X_i}$             |
| 모분산 | $\sigma^2=E[(X-\mu)^2]$|표본분산| $s^2=\frac{1}{n-1}\sum{({X_i}-\bar{X})^2}$|

# 이항분포
> * 확률변수 X가 이항분포를 따를 때, 확률밀도함수는
> * $f(x) = {n \choose k}p^{x}(1-p)^{n-x},\quad x=0,1,...,n$
> * $X \sim B(n,p)$
> * 이항분포의 평균: $\mu=np, \quad 분산: \sigma^2=np(1-p)$

```r
rbinom(n=1, size=10, prob=0.4) # n: 실험을 한번실행, size: 열번시도, prob: 성공확률

binom <- data.frame(Successes = rbinom(n=10000, size=10, prob=0.3))
ggplot(binom, aes(x=Successes)) + geom_histogram(binwidth=1)
```

```r
binom5 <- data.frame(Successes=rbinom(n=10000, size = 5, prob=.3), Size=5)
binom10 <- data.frame(Successes=rbinom(n=10000, size = 10, prob=.3), Size=10)
binom100 <- data.frame(Successes=rbinom(n=10000, size = 100, prob=.3), Size=100)
binom1000 <- data.frame(Successes=rbinom(n=10000, size = 1000, prob=.3), Size=1000)

binomAll <- rbind(binom5, binom10, binom100, binom1000)
dim(binomAll)
head(binomAll, 10)
tail(binomAll, 10)

ggplot(binomAll, aes(x=Successes)) + geom_histogram() + facet_wrap(~Size, scales='free')
# 시도 횟수 크기에 의해 파셋된 임의의 이항 히스토그램
# 시도 횟수가 증가함에 따라 형태가 더 정규화됨
```

# 포아송 분포(Poisson Distribution)
> * X: 단위시간 또는 단위공간에서 발생되는 사건의 수
> * m: 단위시간 또는 단위공간에서 발생되는 사건의 수의 평균, m=E(X)

> * 확률변수 X가 평균 m인 포아송 분포를 따를 때 확률밀도함수는
> * $f(x) = \frac{e^{-m}m^x}{x!}, \quad x=0,1,2,...$
> * $X \sim P(m)$ : Poisson의 P, (테일러급수 참조)
> * 포아송 분포의 평균과 분산 : $\mu=\sigma^2=m$

> * 포아송 분포를 따르기 위한 조건
>> * 독립성(independence): 단위시간 또는 단위공간에서 발생되는 사건의 수는 또 다른 단위시간 또는 단위공간에서 발생되는 사건의 수와 무관함
>> * 단일성(lack of clustrering): 동시에 두 개 이상의 사건이 발생할 확률은 0에 가까움
>> * 등발성(constant rate): 사건의 수 평균 m은 모든 단위시간 또는 단위공간에서 일정

> * 이항분포에서 시행횟수 n이 매우크고 성공의 확률p는 매우작으며, np->m이 되면 이항분포는 포아송 분포와 매우 가까워 짐
> * ${n \choose k}p^{x}(1-p)^{n-x} \approx \frac{e^{-m}m^{x}}{x!}, \quad x=0,1,2,...,n$
> * n이 매우 크고 p가 매우 작다는 것은 사건의 발생이 매우 드물다(rare events)의 뜻

```r
pois1 <- rpois(n=10000, lambda=1)
pois2 <- rpois(n=10000, lambda=2)
pois5 <- rpois(n=10000, lambda=5)
pois10 <- rpois(n=10000, lambda=10)
pois20 <- rpois(n=10000, lambda=20)

pois <- data.frame(Lambda.1 = pois1, Lambda.2=pois2, 
                   Lambda.5 = pois5, Lambda.10=pois10, Lambda.20=pois20)

head(pois)
```

```r
library(reshape2)

pois <- melt(data=pois, variable.name="Lambda", value.name="x")
tail(pois)

library(stringr)
pois$Lambda <- as.factor(as.numeric(str_extract(string=pois$Lambda, pattern="\\d+")))
head(pois)
```

```r
library(ggplot2)

# 람다가 다른 푸아송 분포로부터 1만 개를 추출한 것에 대한 히스토그램
ggplot(pois, aes(x=x)) + geom_histogram(binwidth=1) + 
                       facet_wrap(~ Lambda) + ggtitle("Probability Mass Function")


# 람다가 다른 푸아송 분포로부터 1만 개를 추출한 것의 밀도 plot
ggplot(pois, aes(x=x)) + geom_density(aes(group=Lambda, color=Lambda, fill=Lambda), adjust=4, alpha=1/2) + 
                       scale_color_discrete() + scale_fill_discrete() + 
                       ggtitle("Probability Mass Function")
```

# 정규분포(normal distribution)
> * 정규분포 = 가우스분포(Gaussian distribution)
> * 확률변수 X가 정규분포를 따를 때, 확률밀도함수(pdf)
> <center>$f(x) = \frac{1}{\sqrt{2{\pi}}\sigma}exp\left(-\frac{(x-\mu)^2}{2\sigma^2} \right)$</center>
> * $X \sim N(\mu, \sigma^2)$

> * https://en.wikipedia.org/wiki/Normal_distribution
![](./Empirical_Rule.PNG)

> * 표준정규 분포 : $ Z \sim N(0,1)$
> * $\alpha = P(Z>z_\alpha)$

> * $X \sim N(\mu, \sigma^2)$일 때, 
> * $P(X <=x)$:pnorm($x, \mu, \sigma$) : R code에서 probaibility normal
> * $\alpha = P(X<=x)$를 만족하는 x: qnorm($\alpha, \mu, \sigma$): R code에서 quantile normal, 면적이 어떤 x값에서  $\alpha$가 되는지 알려줌

```r
normal_dist <- rnorm(3000)
normal_density <- dnorm(normal_dist)

library(ggplot2)

ggplot(data.frame(x=normal_dist, y = normal_density)) + 
       aes(x=x, y=y) + geom_point() + labs(x="Random Normal Variables", y= "Density")
```

```r
ggplot(data.frame(x=normal_dist, y = pnorm(normal_dist))) + 
       aes(x=x, y=y) + geom_point() + labs(x="Random Normal Variables", y= "Density")
```

```r
# 정규 분포의 분포를 계산하며 주어진 수 보다 작은수가 발생할 누적확률
pnorm(normal_dist[1:2])
pnorm(c(-3, 0, 3))

# 기본적 좌측 꼬리 검정
pnorm(1) 
pnorm(0)
pnorm(1) - pnorm(0)
```

```r
q <- ggplot(data.frame(x=normal_dist, y = normal_density)) + aes(x=x, y=y)+ 
          geom_line() + labs(x="x", y ="Density")

valueNeg1_seq <- seq(from = min(normal_dist), to = -1, by=.1)
tmp <- data.frame(x=valueNeg1_seq, y=dnorm(valueNeg1_seq))
tmp <- rbind(c(min(normal_dist), 0), tmp, c(max(tmp$x), 0))
head(tmp, 3)
tail(tmp, 3)

q + geom_polygon(data=tmp, aes(x=x, y=y))

valueNeg1Pos1 <- seq(from= -1, to = 1, by=.1)
tmp2 <- data.frame(x=valueNeg1Pos1, y=dnorm(valueNeg1Pos1))
head(tmp2)
tail(tmp2)

tmp2 <- rbind(c(min(tmp2$x), 0), tmp2, c(max(tmp2$x), 0))
q + geom_polygon(data=tmp2, aes(x=x, y=y))
```

# 표준화
> * 확률변수 X가 평균 $\mu$, 분산 $\sigma^2$를 가질 때, 표준화된 변수는 항상 평균 0, 분산 1을 갖음
> * 표준화된 변수(standard random variable): $Z=\frac{X-E(X)}{\sqrt{Var(X)}}=\frac{X-\mu}{\sigma}$

> * 표준화의 개념은 용도가 매우 다양
> * 표준화를 적용할 때 주의사항
>> * 집단의 크기가 적어도 20이상
>> * 각 집단은 동질적이라는 가정

# 표본의 분포
> * 모수(parameter)
>> * 모집단의 특성을 나타내는 값
>> * 알 수 없는 상수인 모수의 값을 추정하기 위해 모집단으로부터 크기 n인 표본을 추출
> * 통계량(statistic)
>> * 표본들만의 함수로 표현 (표본평균, 표본분산)
>> * 통계량은 특정 모수를 추정하기 위해 표본을 이용하여 만든 함수
> * 표본분포(sampling distribution)
>> * 통계량이 갖는 분포
> * 임의표본(random sample)
>> * 크기 n인 표본이 1) 동일한 분포(동일한 모집단)을 가지며, 2) 서로 독립일때 임의표본이라고 함
>> * 모평균이 $\mu$이고, 모분산이 $\sigma^2$인 모집단으로부터 크기 n인 임의표본 $X_1, X_2,..., X_n$을 추출하여 구한 표본평균 $\bar{X}=\frac{1}{n}\sum{X_i}$의 기댓값과 분산은 $E(\bar{X}=\mu), \quad Var(\bar{X})=\frac{\sigma^2}{n}, \quad s.d(\bar{X})=\frac{\sigma}{\sqrt{n}}$
> * 중심극한정리(Central limit therom: CLT)
>> * 표본의 크기 n이 충분히 크면 근사적으로 $ N(\mu, \frac{\sigma^2}{n})$ 분포를 따르게 되며, 이런 사실을 표준화 하면 $Z \equiv \frac{\bar{X}-E(\bar{X})}{s.d(\bar{X}}= \frac{\bar{X}-\mu}{\frac{\sigma}{\sqrt{n}}}=\frac{\sqrt{n}(\bar{X}-E(X_1))}{s.d(X_1}$ 는 근사적으로 N(0,1)을 따르게 됨
>> * 표본의 크기가 클 수록 $\bar{X}$의 분포는 더 정규분포에 가까이 수렴하며, 흔히 표본의 크기가 25보다 크면 중심극한정리를 이용할 수 있음
